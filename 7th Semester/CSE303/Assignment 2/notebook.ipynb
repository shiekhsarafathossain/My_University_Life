Importing Libraries 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Reading and printing the Bike Sales Dashboard Dataset

```python
data = pd.read_excel("Bike Sales Dashboard.xlsx")
print(data)
```

Now I am determining the mean, median, variance and standard deviation of all continuous numerical values.

Income mean

```python
income_mean = data["Income"].mean()
print(income_mean)
```

Income median

```python
income_median = data["Income"].median()
print(income_median)
```

Income variance

```python
income_var = data["Income"].var()
print(income_var)
```

Income standard deviation

```python
income_std = data["Income"].std()
print(income_std)
```

Age mean

```python
age_mean = data["Age"].mean()
print(age_mean)
```

Age median

```python
age_median = data["Age"].median()
print(age_median)
```

Age variance

```python
age_var = data["Age"].var()
print(age_var)
```

Age standard deviation

```python
age_std = data["Age"].std()
print(age_std)
```

Now I am determining the frequency and percentage of each type of all categorical and discrete values.

Children Frequency

```python
children_freq = data["Children"].value_counts()
print(children_freq)
```

Percentage of children type

```python
print(f"{(children_freq/len(data["Children"]))*100}")
```

Cars Frequency

```python
cars_freq = data["Cars"].value_counts()
print(cars_freq)
```

Percentage of cars type

```python
print(f"{(cars_freq/len(data["Cars"]))*100}")
```

Categorical Column: Purchased Bike

```python
purchased_bike = data["Purchased Bike"].value_counts()
print(purchased_bike)
```

Percentage of purchased bike 

```python
print(f"{(purchased_bike/len(data["Purchased Bike"]))*100}")
```

Bar Chart: Frequency of a categorical column (e.g., "Purchased Bike")

```python
# Bar Chart for "Purchased Bike"
plt.figure(figsize=(8, 6))
data['Purchased Bike'].value_counts().plot(kind='bar', color=['skyblue', 'orange'], edgecolor='black')
plt.title('Frequency of Purchased Bike')
plt.xlabel('Purchased Bike')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

Line Graph: Trend of numerical data (e.g., Age vs Income)

```python
plt.figure(figsize=(10, 6))
data.sort_values('Age', inplace=True)
plt.plot(data['Age'], data['Income'], color='blue', alpha=0.7)
plt.title('Income Trend by Age')
plt.xlabel('Age')
plt.ylabel('Income')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

Pie Chart: Distribution of categorical data (e.g., "Commute Distance")

```python
plt.figure(figsize=(8, 8))
data['Commute Distance'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'orange', 'green', 'red', 'purple'])
plt.title('Distribution of Commute Distance')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

Finding missing values

```python

missing_values = data.isnull().sum()


columns_with_missing = missing_values[missing_values > 0]

print("Missing values in each column:")
print(missing_values)

if len(columns_with_missing) > 0:
    print("\nColumns with missing values:")
    print(columns_with_missing)
else:
    print("\nNo missing values in the dataset.")
```

Correlation heat-map with each column.

```python
file_path = 'Bike Sales Dashboard.xlsx'
data = pd.read_excel(file_path, sheet_name='bike_buyers')

numerical_data = data.select_dtypes(include=['number'])

correlation_matrix = numerical_data.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, )
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```
